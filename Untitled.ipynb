```python
import os
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
```

```python
from keras.optimizers import Adam
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#define the base directory
base_dir = r'C:\Users\barra\Desktop\malaria_ML'  # Use raw string to avoid escaping
```

```python
train_dir = os.path.join(base_dir, 'cell_images')
parasitized_dir = os.path.join(train_dir, 'Parasitized')
uninfected_dir = os.path.join(train_dir, 'Uninfected')
```

```python
if not os.path.exists(train_dir):
    raise FileNotFoundError(f"Directory '{train_dir}' does not exist.")
```

```python
#set up image data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2 #splitting validation
)
```

```python
#create generators for training and validating
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',
    subset='training',
    shuffle=True
)
```

```python
if not os.path.exists(parasitized_dir):
    raise FileNotFoundError(f"Directory '{parasitized_dir}' does not exist.")
if not os.path.exists(uninfected_dir):
    raise FileNotFoundError(f"Directory '{uninfected_dir}' does not exist.")
```

```python
val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=32,
    class_mode='binary',
    subset='validation',
)
```

```python
model = Sequential([
    Input(shape=(150,150,3)),
    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_generator, validation_data=val_generator,epochs=10)
```

```python
#plot training loss and accuracy
def plot_training_history(history):
    plt.figure(figsize=(12, 4))
    
    #potting the accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    #plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='val Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show
    
plot_training_history(history)
```

```python
# Evaluate the model
def evaluate_model(model, train_generator):
    # Predict using the generator
    Y_pred = model.predict(train_generator, steps=len(train_generator), verbose=1)
    
    # Convert predictions to binary (1 for parasitized, 0 for uninfected)
    y_pred = np.where(Y_pred > 0.5, 1, 0).flatten()

    # Confusion matrix
    cm = confusion_matrix(train_generator.classes, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Uninfected', 'Parasitized'], yticklabels=['Uninfected', 'Parasitized'])
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title('Confusion Matrix')
    plt.show()

    # Classification report
    print(classification_report(train_generator.classes, y_pred, target_names=['Uninfected', 'Parasitized']))

# Evaluate the model using the training generator
evaluate_model(model, train_generator)
```

```python
model.save('malaria_detection_model.h5')
print("the model has been saved as'malaria_mdetection_model.h5'")
```
